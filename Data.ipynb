# STEP 1

```python
df.state.value_counts(normalize=True)
i=df.state.mode()[0]
i
```

```python
df.state.fillna(i,inplace=True) #cleaning state column
```

```python
df
```

# STEP 2

```python
df.drop('comments',axis=1,inplace=True) #drop comments columns
df.drop('leave',axis=1,inplace=True)
df
```

# STEP 3

```python
df.self_employed.value_counts(normalize=True)
i=df.self_employed.mode()[0]
i
```

```python
df.self_employed.fillna(i,inplace=True) #cleaning self_employed
```

```python
df
```

```python
df.isna().sum()
```

# STEP 4

```python
df.work_interfere.value_counts(normalize=True)
i=df.work_interfere.mode()[0]
i
```

```python
df.work_interfere.fillna(i,inplace=True) #cleaning work_interfere
```

```python
df
```

```python
df.isna().sum() #clean data
```

```python
sns.heatmap(df.isnull(), yticklabels=False)
```
